```python
import pandas as pd
```

```python
dataset = pd.read_csv("kidney_disease.csv")
```

```python
dataset.drop(columns="id", inplace=True)
```

```python
dataset
```

```python
from univariate import Univariate
```

```python
quan, qual = Univariate.quanQual(dataset)
```

```python
quan
```

```python
dataset = dataset[quan]
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
descriptive = Univariate.univariate_report(dataset, quan)
```

```python
descriptive
```

```python
lesser, greater = Univariate.check_outliers_column_names(quan, descriptive)
```

```python
lesser
```

```python
greater
```

```python
# As per above analysis, we can use median
from sklearn.impute import SimpleImputer
import numpy as np
im = SimpleImputer(strategy="median",missing_values=np.nan)
im.fit(dataset[quan])
dataset = im.transform(dataset[quan])
```

```python
preprocessing_table = pd.DataFrame(data=dataset, columns=quan)
preprocessing_table
```
