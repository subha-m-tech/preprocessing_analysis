# Scenario 3

```python
dataset = pd.read_csv("Placement_Data_Full_Class.csv")
dataset = dataset.drop(columns='sl_no')
```

```python

# if you want to replace the values with mean then use below lines
dataset["salary"].fillna(dataset["salary"].mean(), inplace=True)
```

```python
dataset
```

```python
#quan, qual = Univariate.quanQual(dataset)
def quanQual(dataset):
        quan = []
        qual = []
        for column_name in dataset.columns:
            if dataset[column_name].dtype == 'O':
                qual.append(column_name)
            else:
                quan.append(column_name)
        return quan, qual
```

```python
quan, qual = quanQual(dataset)
```

```python
quan
```

# how to handle missing values in one or more columns

```python
dataset = pd.read_csv("Placement_Data_Full_Class.csv")
dataset = dataset.drop(columns='sl_no')
```

```python
# how to handle missing values in one or more columns
import numpy as np
from sklearn.impute import SimpleImputer
im = SimpleImputer(missing_values=np.nan, strategy='mean')
im.fit(dataset[quan])
df = im.transform(dataset[quan])
```

```python
df
```

```python
preprocessing_quan_table = pd.DataFrame(columns= quan, data=df)
preprocessing_quan_table
```

```python
cate_table = dataset[qual]
```

```python
df = [preprocessing_quan_table, cate_table]
```

```python
result = pd.concat(df, axis=1)
```

```python
result
```
